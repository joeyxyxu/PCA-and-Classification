```python
from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA()
Y_sklearn = sklearn_pca.fit_transform(X_std)

cum_sum = sklearn_pca.explained_variance_ratio_.cumsum()

sklearn_pca.explained_variance_ratio_[:10].sum()

cum_sum = cum_sum*100

fig, ax = plt.subplots(figsize=(10,10))
plt.plot(range(30), cum_sum, label='Cumulative _Sum_of_Explained _Varaince', color = 'b',alpha=0.5)
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_');
```

```python
# according to the plot, we will choose the first 11 variables, which explain around 95% of the variance.
sklearn_pca = sklearnPCA(n_components=11)
Y_sklearn = sklearn_pca.fit_transform(X_std)

cum_sum = sklearn_pca.explained_variance_ratio_.cumsum()
print(cum_sum)

sklearn_pca.explained_variance_ratio_[:10].sum()

cum_sum = cum_sum*100

fig, ax = plt.subplots(figsize=(10,10))
plt.bar(range(11), cum_sum, label='Cumulative _Sum_of_Explained _Varaince', color = 'b',alpha=0.5)
plt.title("Around 95% of variance is explained by the Fisrt 11 col ")
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_');

# the variance explained by the first 11 col is 96.1366%
```

```python
# check for the data after PCA
print(Y_sklearn)
X_pca = pd.DataFrame(Y_sklearn)
X_pca.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import confusion_matrix
# split data set to train and test(80-20)
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, Y, test_size=0.2)
```

```python
# Train a logistic regression classifier
logreg = linear_model.LogisticRegression(penalty='l2', C=10, multi_class='ovr')
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
# Print out the accuracy of the classifier
CM = confusion_matrix(Y_test, Y_pred)
print("Accuracy of positive class:",CM[1,1]/(CM[1,0]+CM[1,1]))
print("Accuracy of negative class:",CM[0,0]/(CM[0,0]+CM[0,1]))
```

```python
# Implement decision tree
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=4, random_state=0)
clf = clf.fit(X_train, Y_train)

Y_pred = clf.predict(X_test)
# Print out the accuracy of the classifier
CM = confusion_matrix(Y_test, Y_pred)
print("Accuracy of positive class:",CM[1,1]/(CM[1,0]+CM[1,1]))
print("Accuracy of negative class:",CM[0,0]/(CM[0,0]+CM[0,1]))
```

```python
# Implement random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, random_state=0)
rf.fit(X_train, Y_train)

Y_pred = rf.predict(X_test)
# Print out the accuracy of the classifier
CM = confusion_matrix(Y_test, Y_pred)
print("Accuracy of positive class:",CM[1,1]/(CM[1,0]+CM[1,1]))
print("Accuracy of negative class:",CM[0,0]/(CM[0,0]+CM[0,1]))
```
